```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
theme.head()
```

```python
df.head()
```

```python
merged = df.merge(theme, left_on = 'parent_theme', right_on = 'name')
merged.drop(columns='name_y', inplace=True)
merged.head()
```

# what % of all licensed  sets were star wars

```python
# getting is_licensed True values
licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])
licensed.head()
```

```python
# getting the list of Star Wars themed
star_wars = licensed[licensed['parent_theme']=='Star Wars']
star_wars.shape
```

```python
# finding null values in set_num column
merged[merged['set_num'].isnull()]
```

```python
# dividing the 0 index (column) of star_wars by that of licensed to get the %
the_force = int(star_wars.shape[0] / licensed.shape[0]*100)
print(the_force)
```

# in which year was star wars not the most popular theme

```python
# sort the licensed products by year and theme and then grouping them, and sorting them to get the answer. 2017
licensed_sorted = licensed.sort_values('year')
licensed_sorted['count'] = 1
summed_df = licensed_sorted.groupby(['year', 'parent_theme']).sum().reset_index()
summed_df.head()
```

```python
max_df = summed_df.sort_values('count', ascending=False).drop_duplicates(['year'])
max_df.sort_values('year', inplace=True)
max_df.head(50)
```

# breakdown number of sets by year

```python
clean_df = merged[~merged['set_num'].isnull()]
clean_df['count'] = 1
sets_per_year = clean_df.groupby(['year']).sum().reset_index()[['year','count']]
for index,row in sets_per_year.iterrows():
    print(row['year'],row['count'])
```
